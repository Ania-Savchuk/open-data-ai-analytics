## Підсумки перевірки якості даних (Data Quality Report)

Після проведення первинного аналізу набору даних про реєстрацію ТЗ було виявлено наступне:

### 1. Цілісність даних та пропуски
* **Загальна кількість записів:** 315,679 рядків.
* **Найбільша кількість пропусків** спостерігається у колонці `CAPACITY` (26,033) та `FUEL` (10,759). Пропуски в об'ємі двигуна часто корелюють з електромобілями, що є логічним для даного типу ТЗ.
* **Критичні дані:** Поля `BRAND`, `MODEL` та `MAKE_YEAR` не мають пропусків, що дозволяє проводити подальший аналіз популярності марок та віку авто.

### 2. Дублікати
* Виявлено **710 повних дублікатів**. Це свідчить про наявність технічних повторів у вивантаженні, які необхідно видалити перед етапом дослідження (Research).

### 3. Логічні аномалії та валідація
* **MAKE_YEAR:** Аномальних значень року (менше 1900 або більше 2026) не виявлено. Мінімальний рік у реєстрі — 1900, максимальний — 2026.
* **FUEL:** Виявлено 12 унікальних категорій палива. Присутні неінформативні значення, такі як `.` або `ВІДСУТНЄ`, які потребують додаткової обробки.
* **BRAND:** У реєстрі представлено 1047 унікальних марок. Перевірка через переведення в верхній регістр та видалення пробілів показала, що назви брендів уже уніфіковані (кількість залишилася незмінною — 1047).
